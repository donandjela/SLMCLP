```python
import numpy as np
import matplotlib.pyplot as plt

def rastojanje(A,B):
    x1,y1 = A
    x2,y2 = B
    
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)

def mat_rastojanja(niz):
   
    duzina = len(niz)
    M = np.eye(duzina)
    #print(duzina)
    for i in range(duzina):
        for j in range(duzina):
            
            M[i,j] = rastojanje(niz[i],niz[j])
            
    return np.array(M)


def generisi(broj_tacaka):
    
    x = np.random.randint(0,100,broj_tacaka)
    y = np.random.randint(0,100,broj_tacaka)
    
   
   # x = np.random.random_sample(size = broj_tacaka)*100
   #y = np.random.random_sample(size = broj_tacaka)*100
    
    return np.array([(a,b) for a,b in zip(x,y)])


def get_locations(m,t):
    
    sve = list(range(t))
    np.random.shuffle(sve)
    
    
    facilities = np.sort(sve[:m]).tolist()
    customers = np.sort(sve[m:]).tolist()
    
    return (facilities, customers)
    

def separate(facilities,deo):
    
    np.random.shuffle(facilities)
    k = int(np.round(len(facilities)/deo))
    
    I2 = np.sort(facilities[:k]).tolist()
    I1 = np.sort(facilities[k:]).tolist()
   
    return (I1,I2)



def pref(customers,facilities, D):
    x= np.eye(len(customers), len(facilities))
   # print('matrica rastijanja', D)
    D = np.array(D)
    u = np.random.uniform(size = 1)[0]
    for i in range(len(customers)):
        
        
       # print(i+1,"red",D[customers[i]][facilities])
        m = min(D[customers[i]][facilities])
        M = max(D[customers[i]][facilities])
        
       # print("min {0:.2f}".format(m))
       # print("max{0:.2f}".format(M))
        
        for j in range(len(facilities)):
            k = (D[i,j] - m)/(M-m)    
            
            if 0<=u<=k:
                x[i,j] = m + np.sqrt(u*(M-m)*(D[customers[i]][facilities[j]] - m))
            else:
                x[i,j] = M - np.sqrt((M-D[customers[i]][facilities[j]])*(M-m)*(1-u))
                
    return x

def podeli_kupce(tacke, radius,customers, I2):
    
    J1 = []
    J2 = []
    
    for i in customers:
        for j in I2:
            ind = False
            
            if rastojanje(tacke[i], tacke[j]) < radius:
                ind = True
                break
            
        if ind:
            J2.append(i)
        else:
            J1.append(i)
            
    return (J1,J2)
            
               
```

```python

def foo(radius,p,t,m,dat,deo = 10):



    tacke = generisi(t)
    #print(tacke)
    D = mat_rastojanja(tacke)



    facilities, customers = get_locations(m,t)
    #print(facilities)
    #print(customers)
    I1,I2 = separate(facilities,deo)
    I = I1 + I2 # Nadovezem I2 na I1 
    m1 = len(I1) # new facilities
    m2 = len(I2) #existing facilities


    print("=======  I   ======")
    
    print("Ustanove")
    print("nove,stare")
    print(I1,I2)
    print("nove + stare")
    print(I)

    print("==============")

    


    a,b = zip(*tacke[customers])
    c,d = zip(*tacke[I1])
    e,f  = zip(*tacke[I2])

    
    ax = plt.gca()
    ax.axis('equal')

    plt.scatter(a,b,c='red', label = 'Customers')
    plt.scatter(e,f,label = 'I2')
    plt.scatter(c,d,c='g', label = 'I1')


    for i, txt  in enumerate(tacke):
        ax.annotate(i,tacke[i])


    for i in tacke[facilities]:
        circle1 = plt.Circle(tuple(i),radius = 35, color = 'gray', fill = False)

        ax.add_artist(circle1)

    #plt.legend(loc=1)
    #plt.scatter()




    #formiram skupove  J1, J2, J = J1 U J2

    J1,J2 = podeli_kupce(tacke, radius,customers, I2)
    J = J1 + J2

    print("========  J  =======")

    print("Kupci")
    print(customers)
    print("nisu_oplsluzeni,opsluzeni")
    print(J1, J2)
    print("nisu_opsluzeni+opsluzeni")
    print(J)


    print("==================")


    k1 = len(J1) # new facilities
    k2 = len(J2) #existing facilities
    
    
    #print(pref(customers,facilities, D))
    preference = np.argsort(pref(J,I, D)).T

    print('preference \n',preference)
    
    print("==================")
    
    demand = np.random.randint(0,100,len(J))
    print(demand)
    
    
    
    with open(dat,'w') as f:
    
        f.write(str(radius))
        f.write('\n')
        f.write(str(p))
        f.write('\n')
        #f.write(str(t-m))
        #f.write('\n')

        f.write(str(k1))
        f.write('\n')
        f.write(str(k2))
        f.write('\n')


        f.write(str(m1))
        f.write('\n')
        f.write(str(m2))
        f.write('\n')


        f.write('\n')
        for line in tacke:
            f.write(str(line).strip('[]'))
            f.write('\n')


        f.write('\n')
        for line in preference:
            for item in line:
                f.write(str(item)+" ")
        f.write('\n')

        for i in range(len(J)):
            f.write(str(demand[i]))
            f.write(' ')
        f.write('\n')
        f.write('\n')
        ''' f.write(str(customers).replace(', '," ").strip('[]'))
        f.write('\n')

        '''
        
        f.write(str(J).replace(', '," ").strip('[]'))
        f.write('\n')

        #f.write(str(J1).replace(', '," ").strip('[]'))
        #f.write('\n')
        #f.write(str(J2).replace(', '," ").strip('[]').strip(','))
        #f.write('\n')


        f.write(str(I).replace(', '," ").strip('[]'))
        f.write('\n')

        #f.write(str(I1).replace(', '," ").strip('[]'))
        #f.write('\n')
        #f.write(str(I2).replace(', '," ").strip('[]').strip(','))
        #f.write('\n')
        

    mat = []

    print("I(j)")

    for j in range(len(customers)):
       pom = []
       for i in range(len(facilities)):

           if rastojanje(tacke[J[j]],tacke[I[i]])<=radius:
               pom.append(i)
       mat.append(pom)

    print(mat)

    I_j = mat
    mat = []


    print("J(i)") 
    for i in range(len(facilities)):
       pom = []
       for j in range(len(customers)):

           if rastojanje(tacke[J[j]],tacke[I[i]])<=radius:
               pom.append(j)
       mat.append(pom)

    print(mat)
    J_i = mat
            
    return (preference, I_j,J_i)
```
